# This notebook was created for data aggregation for the Statistical view of the page.

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data_original = pd.read_csv('data/data.csv')
```

# Race

```python
new_data_women = data_original['HER RACE / ETHNICITY'].value_counts(dropna = False)
new_data_women = new_data_women.reset_index(level=0)
new_data_women['index'] = new_data_women['index'].fillna("No Information")
new_data_women
```

```python
new_data_men = data_original['HIS RACE/ETHNICITY'].value_counts(dropna = False)
new_data_men.index = pd.Series(new_data_men.index).replace(np.nan, 'No information')
new_data_men
```

```python
new_data_men = new_data_men.reindex(["White", "Black / African American", "Latino/x", "Unknown / unreleased", "Asian",
                                    "Native American / Alaska Native", "Other (see notes)", "Two or more races",
                                    "Native Hawaiian / Other Pacific Islander", "No information"])
new_data_men = pd.concat([new_data_men, pd.Series({"Two or more races,Black / African American,White": '0'})])
new_data_men = new_data_men.reset_index(level=0)
new_data_men
```

```python
new_data_women = new_data_women.rename(columns = {'index': 'race_label', 'HER RACE / ETHNICITY': 'her_race_count'})
new_data_men = new_data_men.rename(columns = {'index': 'his_race', 0: 'his_race_count'})
data = new_data_women.join(new_data_men)
data = data.drop('his_race', axis = 1)
data
```

# Age

```python
age_women = data_original.copy()
age_women['HER AGE'] = age_women['HER AGE'].apply(lambda row: "0-10" if row <= 10 else
                                                  ("10-20" if row <= 20 else
                                                   ("20-30" if row <= 30 else
                                                    ("30-40" if row <= 40 else
                                                    ("40-50" if row <= 50 else
                                                    ("50-60" if row <= 60 else
                                                    ("60-70" if row <= 70 else
                                                    ("70-80" if row <= 80 else
                                                    ("80-90" if row <= 90 else
                                                     ("90+" if row == row else "No information")))))))))) 
age_women = age_women['HER AGE'].value_counts()
age_women = age_women.sort_index()
age_women = age_women.reset_index(level=0)
age_women
```

```python
age_men = data_original.copy()
age_men['HIS AGE'] = age_men['HIS AGE'].apply(lambda row: "0-10" if row <= 10 else
                                                  ("10-20" if row <= 20 else
                                                   ("20-30" if row <= 30 else
                                                    ("30-40" if row <= 40 else
                                                    ("40-50" if row <= 50 else
                                                    ("50-60" if row <= 60 else
                                                    ("60-70" if row <= 70 else
                                                    ("70-80" if row <= 80 else
                                                    ("80-90" if row <= 90 else
                                                     ("90+" if row == row else "No information"))))))))))
age_men = age_men['HIS AGE'].value_counts()
age_men = age_men.sort_index()
age_men = age_men.reset_index(level=0)
age_men
```

```python
age_women['his_age_count'] =  age_men['HIS AGE']
age_women
```

```python
data['age_label'] = age_women['index']
data['her_age_count'] = age_women['HER AGE']
data['his_age_count'] = age_women['his_age_count']
data
```

```python
pd.DataFrame.to_csv(data, 'data/Race_Age.csv')
```

# Relationship

```python
relationship = data_original.copy()
relationship = relationship['RELATIONSHIP'].value_counts(dropna = False)
relationship = relationship.reset_index(level=0)
relationship['index'] = relationship['index'].fillna("No Information")
relationship = relationship.rename(columns = {'index': 'relationship_label', 'RELATIONSHIP': 'relationship_count'})
relationship
```
